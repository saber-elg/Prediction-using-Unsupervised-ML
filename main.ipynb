Finding the optimum number of clusters

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset into a DataFrame
iris = pd.read_csv('Iris.csv', index_col=0)

# Display the first few rows of the DataFrame
print(iris.head(10))
```

```python
# Initialize empty list to store WCSS values
list = []

# Extracting the numeric data (features) from the DataFrame
X = iris.iloc[:, [0, 1, 2, 3]].values  

# Calculate WCSS for k between 1 and 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    list.append(kmeans.inertia_)  

# Plotting the Elbow method graph
plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), list, marker='*', linestyle='solid')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python

# Based on the results from the Elbow Method ,we initialize KMeans with the optimal number of clusters (we choose 3 clusters)
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)

# Fit KMeans to the data
kmeans.fit(X)

# Predict the cluster labels
labels = kmeans.predict(X)

# Add the predicted labels to the original DataFrame
iris['Cluster'] = labels
```

Visualize the clusters

```python
# Plotting the clusters
plt.figure(figsize=(8, 6))

# Scatter plot for Sepal features
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8, edgecolors='k', label='Data Points')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='*', edgecolors='k', linewidths=1, label='Centroids')

plt.title('KMeans Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```
